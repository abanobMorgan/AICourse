```python
baby_names[:10]
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names.drop(columns= "Unnamed: 0", inplace= True)
```

### Step 6. Is there more male or female names in the dataset?

```python
# here we are going to count the males in the data set then compare them with 
#the females in the same data set if the females less than the males we will peint True other wise 
# we will print Flase 
males = baby_names[baby_names['Gender'] == 'M'].count()[0]
females = baby_names[baby_names['Gender'] == 'F'].count()[0]
print(f'the males number is:\t{males} \nthe females number is:\t {females}  ')
males > females
```

### Step 7. Group the dataset by name and assign to names

```python
baby_names.columns
```

```python
names = baby_names.groupby("Name")
names.count()
```

### Step 8. How many different names exist in the dataset?

```python
names.count().count()[0]
```

### Step 9. What is the name with most occurrences?

```python
names.count().max()[0]
```

```python
names.count().index[names.count().max()[0]]
```

### Step 10. How many different names have the least occurrences?

```python
newnames =  names.count().values[:, 0]
len(newnames)
```

```python
counter_occ = 0
for i in range(len(newnames)):
    if newnames[i] == 1 : 
        print(f'the value of row is: {i} and the value of counter is {counter_occ}. ')
        counter_occ +=1 
print(counter_occ)
```

### Step 11. What is the median name occurrence?

```python
occ= baby_names.groupby("Name")['Id'].count().reset_index(name='count').sort_values(['count'])
occ['Name'][len(newnames)//2]
```
